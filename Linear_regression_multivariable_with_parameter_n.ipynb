```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('C:\\Users\\PRASHANT\\Desktop\\coding dec19\\weather-data-in-new-york-city-2016\\ex1data2.csv')
df1.head()
```

```python
#x1=df1.iloc[:,[0]]
#x2=df1.iloc[:,[1]]
m,n=df1.shape
XX=df1.iloc[:,0:n-1]
print(XX.shape,type(XX))
yy=df1.iloc[:,[n-1]]
print(yy.shape,type(yy))

print("m= {}, n={} ".format(m,n))
n=n-1
#print(X.iloc[:,[0]])
print("updated m= {}, n={} ".format(m,n))
```

```python
x0=np.ones([m,1])
X=x0
print(X.shape)
```

```python
X=np.ones([m,1])
div_normaliser=[]
sub_normaliser=[]


for j in range(0,n):   # nth column will not be included there are n column in X 0 to n-1
    x1=df1.iloc[:,[j]]
    #print(x1)
    x1=np.array(x1)
    #print(x1)
    sub_normaliser.append(x1.mean()) #soring for further
    print("x1 mean is ",x1.mean())
    x11=np.ones([m,1])
    x11=x11*x1.mean()
    
   
    div_normaliser.append(x1.std())#storing for further calculation
    print("x1 standard deviation is ",x1.std())
    x1=(x1-x11)/x1.std()
     
    
    print(x1.shape,type(x1))
    X=np.hstack([X,x1])   #one by one add all row
#print(X)
print(X.shape,type(X))
#x1=x1-(np.ones([46,1]))* x1.mean
#print(x11)
#print(x1)

```

```python
Y=np.array(yy)
print(Y.shape,type(Y))
print(X.shape,type(X))
```

```python
theta=np.zeros([n+1,1])
alpha=0.01
print(" alpha ie learning rate is ",alpha)
print("theta is \n",theta)
```

```python
X2=X.dot(theta)-Y
X3=X2**2 #perform element wise square
cost=(1/(2*m))*np.sum(X3)
cost=float(cost)
print(cost)
```

```python

lx=[]
lx.append(0)
ly=[]
ly.append(cost)
for i1 in range(15000):
    X2=X.dot(theta)-Y
    #X4=(1/m)*np.sum(X2)*alpha
    #X9=np.array(X9)
    #X5=X2*X9
    #X6=(1/m)*np.sum(X5)*alpha
    
    X4=(1/m)*(X.transpose()).dot(X2)
    #theta[0]=theta[0]-X4
    #theta[1]=theta[1]-X6
    #calculating cost after each iteration
    theta=theta-alpha*X4
    
    if i1%50==0:
        X2=X.dot(theta)-Y
        X3=X2**2 #perform element wise square
        cost=(1/(2*m))*np.sum(X3)
        cost=float(cost)
        print("i1 = {}, cost = {} ".format(i1,cost))
        lx.append(i1*5)
        ly.append(cost)
       # 
       
    
```
